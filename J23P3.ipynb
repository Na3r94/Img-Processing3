```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('Mona_Lisa.jpg')
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
new_image = np.zeros(image.shape, image.dtype)
alpha = 1.0 # Simple contrast control
beta = 0    # Simple brightness control

print(' Basic Linear Transforms ')
try:
    alpha = float(input('* Contrast control - Enter the alpha value [1.0-3.0]: '))
    beta = int(input('* Brightness control - Enter the beta value [0-100]: '))
except ValueError:
    print('Error, not a number')

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        new_image[i, j] = np.clip(alpha*image[i, j] + beta, 0, 255)

plt.imshow(image, cmap='gray')
```
